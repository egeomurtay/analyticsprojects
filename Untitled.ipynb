```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from scipy.stats import zscore
import numpy as np
from sklearn.metrics import r2_score,mean_squared_error
```

```python
surface_df=pd.read_csv('SurfaceFinish.csv')
```

```python
surface_df.head(5)
```

```python
surface_df.columns=['Part','Surface_Finish','RPM','Cutting_Tool']
```

```python
surface_df.head(12)
```

```python
X_features = ['RPM', 'Cutting_Tool'] 
X_df = surface_df[X_features]
X = sm.add_constant(X_df)
Y = surface_df['Surface_Finish']
```

```python
surface_dummy = pd.get_dummies(surface_df['Cutting_Tool'], prefix='Cutting_Tool',drop_first=True)
surface_dummy.head()
```

```python
X=X.drop('Cutting_Tool',axis = 1)
RevisedX = pd.concat([X, surface_dummy], axis=1)
RevisedX.head()
```

```python
surface_lm = sm.OLS(Y,RevisedX).fit()
surface_lm.summary2()
```

```python
beverage_df=pd.read_csv('BeverageSales.csv')
```

```python
beverage_df.head()
```

```python
beverage_df.columns=['Temp','$Sales']
```

```python
beverage_df.head()
```

```python
plt.scatter(x=beverage_df['Temp'],y=beverage_df['$Sales']);
plt.xlabel('Temprature')
plt.ylabel('Sales in $')
plt.show()
```

```python
X_features = ['Temp'] 
X_df = beverage_df[X_features]
X = sm.add_constant(X_df)
Y = beverage_df['$Sales']
```

```python
bev_model = sm.OLS(Y, X).fit()
bev_model.summary2()
```

```python
plt.scatter(zscore(bev_model.fittedvalues),zscore(bev_model.resid))
plt.title('Residual Plot:Sales Prediction');
plt.xlabel('Standardized predicted values')
plt.ylabel('Standardized residuals');

#residuals got funnel shape linear rship aint appropriate
```
