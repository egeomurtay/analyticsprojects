```python
#j) Derive a new column "agegroup" from age column
ageCategorical=list(range(len(heart_df.age)))
```

```python
for i in range(len(heart_df.age)):
    if heart_df.iloc[i][8]<=18:
        ageCategorical[i] = 'young'
    elif heart_df.iloc[i][8]>18 and heart_df.iloc[i][8]<=35:
        ageCategorical[i] = 'adult'
    elif heart_df.iloc[i][8]>35 and heart_df.iloc[i][8]<=55:
         ageCategorical[i] = 'mid'
    else:
        ageCategorical[i] = 'old'   
```

```python
heart_df['agegroup'] = ageCategorical
```
